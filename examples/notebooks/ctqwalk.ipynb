# Continuous-Time Quantum Walks

### Adjacency Matrix of K3

```python
import numpy as np
from walkg.quantum import CTQW

K3 = np.array([[0, 1, 1], [1, 0, 1], [1, 1, 0]])
```

### Quantum Walk on K3

```python
w = CTQW(K3)
```

### Transition Matrix

```python
# PST
np.round(w.transition_matrix(2*np.pi/3),3)
```

```python
# Instantaneous Mixing
np.round(w.transition_matrix(2*np.pi/9),3)
```

### Mixing Matrix

```python
# PST
np.round(w.mixing_matrix(2*np.pi/3),3)
```

```python
# Instantaneous Mixing
np.round(w.mixing_matrix(2*np.pi/9),3)
```

### Average Mixing Matrix

```python
w.avg_matrix()
```

### Generalized Average Mixing Matrix

```python
np.random.seed(10)
samples = np.random.beta(5, 1, size=10000)
w.gavg_matrix(samples)
```

### Density Plot

```python
w.plot_density(2, 3)
```

```python
w.plot_density(1, 1, t_init=0, t_final=7)
```
